```python
# myList = SLL()
# for i in range(5):
#     mylist.append(i)
    
# current = mylist.head
# for i in range(5):
#     assert current.data = i
    
# mylist.insert(0, -1)
# assert mylist.head.data == -1

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

        
class SLL:
    def __init__(self):
        self.head = None
        
    def append(self, item):
        # if list is empty
        if not self.head:
            self.head = Node(item)
        
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = Node(item)
            
    def insert(self, index, item):
        new_node = Node(item)
        if not self.head:
            self.head = new_node
            
        elif index == 0:
            new_node.next = self.head
            self.head = new_node
            
        else:
            counter = 0
            current = self.head
            prev = None
            while current.next and index != counter:
                prev = current.next
                counter += 1
                
            if index == counter:
                prev.next = new_node
                new_node.next = current
            else:
                current.next = new_node
                
                
    def remove(self, value):
        current = self.head
        if current.data == value:
            self.head = current.next
            
        else:
            prev = None
            while current.next and current.data != value:
                prev = current
                current = current.next
                
            if current.data == value:
                prev.next = current.next
                
            else:
                raise ValueError(f'{value} is not in the list')
    
    def __str__(self):
        output_str = "["
        current = self.head
        output_str += f"{current.data}, "
        while current:
            output_str += f",  {current.data}"
        output_str += "]"
        return output_str
        
        
        
mylist = SLL()


for i in range(5):
    new_node = Node(i)
    mylist.append(new_node)
print(mylist)   
current = mylist.head

for i in range(5):
    assert current.data == i
    current = current.next
mylist.insert(0, -1)
assert mylist.head.data == -1


    
current = mylist.head
for i in range(5):
    assert current.data == i

    
mylist.insert(0, -1)
assert mylist.head.data == -1
        
```
